```python
from datasets import load_dataset_builder, load_dataset
from datasets import get_dataset_config_names

from tqdm.auto import tqdm  # for our loading bar
from sumy.parsers.plaintext import PlaintextParser
from sumy.nlp.tokenizers import Tokenizer
from sumy.summarizers.lex_rank import LexRankSummarizer
import pandas as pd
import nltk
nltk.download('punkt')

from google.colab import drive
drive.mount('drive')

def get_num_words(sentence):
  print(len(sentence.split()))
  return len(sentence.split())
```

```python
# PUBMED
configs = get_dataset_config_names("ccdv/pubmed-summarization")
print(configs)

ds_builder = load_dataset_builder("ccdv/pubmed-summarization", "document")
print(ds_builder.info.description)
print(ds_builder.info.features)

dataset = load_dataset("ccdv/pubmed-summarization")
print(dataset)
print(dataset['train'][1])

small_train_dataset = dataset["train"].shuffle(seed=42).select(range(1000))
small_eval_dataset = dataset["validation"].shuffle(seed=42).select(range(100))

```

```python
# ARXIV
configs = get_dataset_config_names("ccdv/arxiv-summarization")
print(configs)

ds_builder = load_dataset_builder("ccdv/arxiv-summarization", "document")
print(ds_builder.info.description)
print(ds_builder.info.features)

dataset = load_dataset("ccdv/arxiv-summarization")
print(dataset)
print(dataset['train'][1])

small_train_dataset = dataset["train"].shuffle(seed=42).select(range(1000))
small_eval_dataset = dataset["validation"].shuffle(seed=42).select(range(100))

```

```python
from datasets import load_metric

metric = load_metric("rouge")


batch_count = 0
batches = []
lex_rank_summarizer = LexRankSummarizer()
print(small_eval_dataset[0])
index = 0

text_data = [] #articles
summaries = []
abstracts = []
length_article = []
length_summary = []
length_abstract = []
rouge_scores = []
data = []
for sample in tqdm(small_eval_dataset):
    index+=1
    # remove newline characters from each sample as we need to use exclusively as seperators
    print("abstract")
    abstract = sample['abstract']
    print(abstract)
    abstracts.append(abstract)
    length_abstract.append(get_num_words(abstract))

    sample = sample['article'].replace('\n', '')
    parser = PlaintextParser.from_string(sample, Tokenizer("english"))
    lexrank_summary = lex_rank_summarizer(parser.document,sentences_count=5)

    text_data.append(sample)
    length_article.append(get_num_words(sample))
    print("Lex Rank Summary")
    summary = ""
    for sentence in lexrank_summary:
      print(sentence)
      summary += str(sentence)
    summaries.append(summary)
    length_summary.append(get_num_words(summary))
    data.append([summary, sample, abstract, get_num_words(summary), get_num_words(sample), get_num_words(abstract)])

baseline_rouge = metric.compute(predictions=summaries, references=abstracts)

output = pd.DataFrame(data, columns = ['Summary', 'Article', 'Abstract', 'Summary Length', 'Article Length', 'Abstract Length'])
#output2 = pd.DataFrame({'Summary': summaries, 'Article': text_data, 'Abstract': abstracts, 'Summary Length': length_summary, 'Article Length': length_article,'Abstract Length': length_abstract})

print("HELLO")
output['Summary'][0]
output['Abstract'][0]
output
```
